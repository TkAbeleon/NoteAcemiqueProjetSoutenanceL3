Résoudre le système d’équation suivante :

$$x \equiv 3 (mod 7)$$
$$x \equiv 1 (mod 11)$$
$$x \equiv 5 (mod 13)$$
$$x \equiv 15 (mof 17)$$
$$x \equiv 12 (mod 23)$$
$$x = 79201$$


```python
def modulo(a: int,b: int, c: int):
    print(f"- {a} = {a%b} mod{b}---{c}")
    return a%b
```

```python
modulo(5, 4,3)
```

```python
x = 0
while False:
    print("***************")
    print(f"\tx = {x}")
    a = modulo(x, 7, 3)
    b = modulo(x, 11, 1)
    c = modulo(x, 13, 5)
    d = modulo(x, 17, 15)
    e =modulo(x, 23, 12)
    if a==3 and b==1 and c==5 and d==15 and e==12:
        break

    x+=1
print(f"$$$$$$$$$$$$$$$$$$$$$$$$$$")
print(f"x = {x}")
```
